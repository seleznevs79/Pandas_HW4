```python
import matplotlib.pyplot as plt
df = pd.read_csv('kc_house_data.csv', encoding = 'latin-1')
df.head()
```

Большая часть недвижисоти продается в диаазоне до 1 млн.

```python
# 1.3. Изучите распределение квадратуры жилой
sns.histplot(df['sqft_living'], bins=20)
plt.title('Распределение жилой площади')
plt.xlabel('sq_ft')
plt.ylabel('Количество')
plt.grid()
```

Наибольшее количество домов (около 11 тысяч) имеют жилую площадь около 2000 кв. футов.  
7,5 тыс домов имеют жилую площадь около 1300 тыс футов.

```python
# 1.4. Изучите распределение года постройки
sns.histplot(df['yr_built'])
plt.title('Распределение по годам постройки')
plt.xlabel('года')
plt.ylabel('Количество')
```

Начиная с начала 20го века идет плавный рост количества представленных домов по годам постройки. Провалы тренда по годам наблюдаются  в 1937г, 1976г, 1997г и в 2010г

```python
# 2.1. Изучите распределение домов от наличия вида на набережную
data = df['waterfront'].value_counts()
plt.pie(data.values, labels=data.index, autopct='%1.2f%%')
plt.legend(data.index)
plt.title('Вид на набережную')
```

Меньше 1% домов имеют вид на набережную

```python
# 2.2. Изучите распределение этажей домов

data=df['floors'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(data.values, labels=data.index, autopct='%1.1f%%')
plt.legend(data.index)
plt.title('Распределение этажей')
```

Подавляющее большинство домов с 1м или 2мя этажами

```python
# 2.3. Изучите распределение состояния домов
data = df['condition'].value_counts()
# plt.bar(data.index, data.values)
plt.figure(figsize=(8, 8))
plt.pie(data.values, labels=data.index, autopct='%1.1f%%')
plt.legend(data.index)
plt.title('Состояние домов')
```
